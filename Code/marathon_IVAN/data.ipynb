```python
import pandas as pd 

df = pd.read_csv('Data.csv')
```

```python
# Display the first few rows of the dataframe
df.head()
```

```python
df.info()
```

```python
# Columns to drop
columns_to_drop = ['Indic4', 'Indic7', 'Indic17']

# Drop the specified columns
df.drop(columns=columns_to_drop, inplace=True)

# Drop rows with any remaining missing values
df.dropna(inplace=True)

# Display the first few rows of the cleaned dataframe
df.head()
```

```python
# Assuming `df` is your dataframe after cleaning and loading

# Attempt to convert all columns to floats
for col in df.columns:
    try:
        df[col] = df[col].astype(float)
    except ValueError:
        # If conversion fails, print the column name
        print(f"Column {col} can't be converted to float.")
        
# After conversion, check the data types to confirm
df.dtypes
```

```python
df.to_csv('data_test.csv', index=False)
```
