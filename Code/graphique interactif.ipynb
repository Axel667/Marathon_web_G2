```python
import pandas as pd
import plotly.express as px
from ipywidgets import interact, Dropdown
from google.colab import files

# Téléchargement et chargement des données
uploaded = files.upload()
data = pd.read_excel(next(iter(uploaded.keys())))
```

```python
# Nettoyer et préparer les données pour la visualisation
# Supprimer les lignes et colonnes inutiles

# Suppression des premières lignes et de la colonne inutile
cleaned_data = data.iloc[4:].reset_index(drop=True)

# Renommer les colonnes
columns = [
    'Code_commune', 'Nom_commune', 'Densite_population_2020', 'Potentiel_financier_2023',
    'Revenu_disponible_median_2020', 'Part_menages_voiture', 'Part_menages_monoparentaux',
    'Part_personnes_75_ans_seules', 'Taux_proprete', 'Taux_equipement_internet',
    'Indice_vieillissement_2020', 'Temps_acces_chef_lieu_proximite', 'Temps_acces_bibliotheques',
    'Temps_acces_colleges', 'Part_cadres_professions_intellectuelles', 'Part_ouvriers',
    'Taux_chomage_15_ans_et_plus', 'Densite_logements_sociaux'
]

cleaned_data.columns = columns

# Convertir les données numériques
numeric_columns = columns[2:]  # Toutes les colonnes après 'Nom_commune' sont numériques
cleaned_data[numeric_columns] = cleaned_data[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Afficher les premières lignes pour vérifier la structure des données après nettoyage
cleaned_data.head()
```

```python
# @title Revenu_disponible_median_2020

from matplotlib import pyplot as plt
cleaned_data['Revenu_disponible_median_2020'].plot(kind='hist', bins=20, title='Revenu_disponible_median_2020')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
import plotly.express as px


median_revenue = cleaned_data['Revenu_disponible_median_2020'].median()
cleaned_data['Revenu_disponible_median_2020'].fillna(median_revenue, inplace=True)

fig = px.scatter(
    cleaned_data,
    x="Densite_population_2020",
    y="Potentiel_financier_2023",
    size="Revenu_disponible_median_2020",
    color="Nom_commune",
    hover_name="Nom_commune",
    size_max=60,
    title="Relation entre la densité de population, le potentiel financier et le revenu disponible médian par commune"
)

fig.update_layout(
    xaxis_title="Densité de population 2020",
    yaxis_title="Potentiel financier 2023",
    legend_title="Commune"
)


fig.show()
```
